```python
## import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=pd.read_csv('covid19_india.csv')
data
```

```python
data=data.iloc[:,1:]
data
```

```python
day=data[(data['Current Status']=="Hospitalized")].groupby(['Month','Day'])[['Num Cases']].sum()
```

```python
x=np.arange(len(day))
x
```

```python
y=day.values
```

```python
x=x.reshape(-1,1)
```

```python
plt.scatter(x,y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly=PolynomialFeatures(degree=2)
```

```python
X=poly.fit_transform(x)
```

```python
pd.DataFrame(X)
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
```

```python
reg.fit(X,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
yp=reg.predict(X)
yp
```

```python
plt.scatter(x,y)
plt.plot(x,yp)
```

```python
reg.score(X,y)*100
```

```python
reg.predict(poly.transform([[301]]))
```
