```python
## import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=pd.read_csv('covid19_india.csv')
data
```

```python
data=data.iloc[:,1:]
data
```

```python
day=data[(data['Current Status']=="Hospitalized")].groupby(['Month','Day'])[['Num Cases']].sum()
```

```python
x=np.arange(len(day))
x=x.reshape(-1,1)
```

```python
y=day.values
```

```python
from sklearn.tree import DecisionTreeRegressor
reg=DecisionTreeRegressor()
```

```python
reg.fit(x,y)
```

```python
Yp=reg.predict(x)
```

```python
plt.scatter(x,y)
plt.plot(x,Yp)
plt.show()
```

```python
reg.score(x,y)*100
```

```python
reg.predict([[100]])
```
