```python
## import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=pd.read_csv('covid19_india.csv')
data.head()
```

```python
data=data.iloc[:,1:]
data
```

```python
day=data[(data['Current Status']=="Hospitalized") & (data['Year']==2020.0)].groupby(['Month','Day'])[['Num Cases']].sum()
```

```python
len(day)
```

```python
x=np.arange(len(day))
x
```

```python
y=day.values
y
```

```python
plt.scatter(x,y)
```

```python
x=x.reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
```

```python
reg.fit(x,y)
```

```python
y[121]
```

```python
y[122]
```

```python
reg.predict([[122]])
```

```python
y_pred=reg.predict(x)
```

```python
y_pred
```

```python
plt.scatter(x,y)
plt.plot(x,y_pred)
```

```python
reg.score(x,y)*100
```
