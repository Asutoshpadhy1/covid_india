```python
## import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=pd.read_csv('covid19_india.csv')
data.head()
```

```python
data=data.iloc[:,1:]
data
```

```python
day=data[(data['Current Status']=="Hospitalized") & (data['Year']==2020.0)].groupby(['Month','Day'])[['Num Cases']].sum()
```

```python
day
```

```python
x=np.arange(len(day))
```

```python
x
```

```python
y=day.values
```

```python
y
```

```python
x=x.reshape(-1,1)
y=y.reshape(-1,1)
```

```python
from sklearn.preprocessing import StandardScaler
sc_X=StandardScaler()
sc_Y=StandardScaler()
Sx=sc_X.fit_transform(x)
Sy=sc_Y.fit_transform(y)
```

```python
from sklearn.svm import SVR
reg=SVR(kernel='rbf')
reg.fit(Sx,Sy.ravel())
```

```python
reg.score(Sx,Sy)*100
```

```python
plt.scatter(Sx,Sy)
plt.plot(Sx,reg.predict(Sx),color='k')
```
